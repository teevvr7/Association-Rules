### Step 1:  turn the table into - transaction(row),products(column),value(quantity)

```python
uk_pivotted_data=pd.pivot_table(data= uk_data,values='Quantity',index='InvoiceNo',columns='Description',aggfunc='sum').fillna(0)
```

```python
uk_pivotted_data
```

### Step 2: change that table into 0,1 dataframe (one-hot df)

```python
import pandas as pd
print(pd.__version__)
```

```python
uk_pivotted_onehotdata=uk_pivotted_data.map(lambda x:0  if x==0 else 1)
# uk_pivotted_onehotdata = uk_pivotted_data.applymap(lambda x: 0 if x == 0 else 1)


# apply() - function only can apply to a column
# applymap() - function can apply to whole dataframe but it is in older version
# better use map()
```

## 4. Data Mining

```python
from mlxtend.frequent_patterns import apriori,association_rules 
#apriori - to build best frequent set-items
#association rules - to build best association path (antecedent , consequent)
```

```python
apriori_data = apriori(df=uk_pivotted_onehotdata,min_support=0.02,use_colnames=True) # df receive one-hot dataframe
apriori_data
# so, turn the table into - transaction(row),products(column),value(quantity)
# then, change that table into 0,1 dataframe
```

```python
best_associations =association_rules(df=apriori_data,metric='confidence',min_threshold=0.02,)
best_associations
```

### Lift
Lift = 1 (No Connection): Buying Item A has zero effect on Item B. It’s pure coincidence. (Example: Buying Milk and buying a Toothbrush. People need both, but buying milk doesn't make you suddenly want to brush your teeth).

Lift > 1 (Positive Connection): Buying Item A boosts the chances of buying Item B. They love each other! (Example: Burgers and Buns. If the Lift is 3, it means a customer is 3 times more likely to buy Buns if they already have Burgers in their cart).

Lift < 1 (Negative Connection): Buying Item A actually lowers the chances of buying Item B. They hate each other! (Example: If someone buys Coke, they are less likely to buy Pepsi at the same time).

A Simple Example (Let's do the Math!)Imagine we look at 100 shopping carts in your store:

50 carts have Burgers in them (50%).

40 carts have Buns in them (40%).

30 carts have Both Burgers and Buns (30%).

If Burgers and Buns had absolutely nothing to do with each other, we would expect them to end up in the same cart purely by chance. To find that "chance," we multiply their individual probabilities:$$\text{Chance} = 50\% \times 40\% = 20\%$$(So, by pure luck, we'd expect 20 out of 100 carts to have both).But in reality, 30% of carts have both!To find the Lift, we divide the real world by pure chance:$$\text{Lift} = \frac{\text{Real World Togetherness}}{\text{Pure Chance Togetherness}} = \frac{30\%}{20\%} = 1.5$$The Result:Because 1.5 is greater than 1, we have proven that Burgers and Buns are a great match. Specifically, having Burgers in the cart lifts the chances of buying Buns by 1.5 times (or 50% more than usual).
